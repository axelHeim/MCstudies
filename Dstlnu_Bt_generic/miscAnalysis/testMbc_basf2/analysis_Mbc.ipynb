```python
fileY4S = uproot.open(root_path + "DXtagDstl" + identifier + ".root:variables")
df_Y4S = fileY4S.arrays(library="pd")
```

```python
names = ["gammas","electrons","pions","kaons","muons"]
dfs = []
for name in names:
    filename = root_path + "{}".format(name) + identifier + ".root:variables"
    print(filename)
    tmpFileFSPs = uproot.open(filename)
    df_tmp = tmpFileFSPs.arrays(library="pd")
    dfs.append(df_tmp)
df_FSPs = pd.concat(dfs)
```

```python
df_Y4S['Hc_px']=df_Y4S['Hc_Px']
df_Y4S['Hc_py']=df_Y4S['Hc_Py']
df_Y4S['Hc_pz']=df_Y4S['Hc_Pz']
```

```python
df_FSPs[df_FSPs["basf2_X"]==1]
```

```python
predCase = [1,"X"]
df_predCase = df_FSPs[df_FSPs["basf2_X"]==1]

fourMom_summed_preds = tmp = pd.DataFrame({f'px_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["px"].sum(),
                             f'py_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["py"].sum(),
                             f'pz_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["pz"].sum(),
                             f'E_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["E"].sum(),
                       
                       f'CMSpx_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["cmpx"].sum(),
                             f'CMSpy_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["cmpy"].sum(),
                             f'CMSpz_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["cmpz"].sum(),
                             f'CMSE_summed_{predCase[1]}' : df_predCase.groupby(["__event__"])["cmE"].sum()}).reset_index()
```

```python
fourMom_summed_preds = pd.merge(fourMom_summed_preds,df_Y4S[["__event__","Hc_px","Hc_py","Hc_pz","Hc_E","dau0_Mbc","dau0_cmpx","Hc_cmPx","dau0_cmpy","Hc_cmPy","dau0_cmpz","Hc_cmPz"]],on=["__event__"])
```

```python
fourMom_summed_preds["Mbc_Btag"] = fourMom_summed_preds.apply(Mbc_Btag,axis=1)
```

```python
fourMom_summed_preds["Btag_cm_px"] = fourMom_summed_preds["Hc_cmPx"] + fourMom_summed_preds["CMSpx_summed_X"]
fourMom_summed_preds["Btag_cm_py"] = fourMom_summed_preds["Hc_cmPy"] + fourMom_summed_preds["CMSpy_summed_X"]
fourMom_summed_preds["Btag_cm_pz"] = fourMom_summed_preds["Hc_cmPz"] + fourMom_summed_preds["CMSpz_summed_X"]
```
